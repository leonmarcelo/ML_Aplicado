##Â Needed Libraries

```python
from sklearn import datasets

from sklearn.cluster import KMeans 

from matplotlib.image import imread
from matplotlib import pyplot as plt
import os
```

## Loading iris dataset:

```python
iris = datasets.load_iris()
```

```python
list(iris.keys())
```

```python
X = iris["data"][:,0:2]
y = iris["target"]
```

```python
print("Data:\n", X[0:10,:])
print("\ntarget (we will not use the labels!)\n",y)
```

## K-Means

```python
k=3
kmeans = KMeans(n_clusters=k) 
y_pred = kmeans.fit_predict(X)
```

```python
y_pred
```

```python
kmeans.cluster_centers_
```

```python

 
#filter rows of original data
data_label0 = X[y_pred==0]
data_label1 = X[y_pred == 1]
data_label2 = X[y_pred == 2]

 
#plotting the results
plt.figure()
plt.scatter(data_label0[:,0] , data_label0[:,1],label="Cluster 0")

#filter rows of original data

 
#Plotting the results
plt.scatter(data_label1[:,0] , data_label1[:,1] , color = 'cyan', label="Cluster 1")
plt.scatter(data_label2[:,0] , data_label2[:,1] , color = 'red', label="Cluster 2")
plt.legend(loc="best")
plt.title("Iris dataset, clustering")
plt.show()

plt.show()
```

```python
#filter rows of original data
label0 = X[y == 0]
label1 = X[y == 1]
label2 = X[y == 2]

#plotting the results
plt.figure()
plt.scatter(label0[:,0] , label0[:,1],label="Class 0")
plt.scatter(label1[:,0] , label1[:,1] , color = 'cyan', label="Class 1")
plt.scatter(label2[:,0] , label2[:,1] , color = 'red', label="Class 1")
plt.legend(loc="best")
plt.show()

plt.show()

```

# K-Means for image segmentation
*****

```python
# Reading the image
image = imread(os.path.join("images","ladybug.png"))
```

```python
image.shape
```

```python
plt.figure()
plt.imshow(image)
plt.show()
```

```python
# Reshaping for getting a matrix of mxn, each row is a pixel 
# represented by a vector of 3 components
X = image.reshape(-1, 3)
print(X.shape)
print(X)
```

```python
# from https://stackoverflow.com/questions/51245877/image-segregation-by-k-mean-python
kmeans = KMeans(n_clusters=8, random_state=42).fit(X)
segmented_img = kmeans.cluster_centers_[kmeans.labels_]
segmented_img = segmented_img.reshape(image.shape)
segmented_imgs = []

n_colors = (10, 8, 6, 4, 2)
for n_clusters in n_colors:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42).fit(X)
    segmented_img = kmeans.cluster_centers_[kmeans.labels_]
    segmented_imgs.append(segmented_img.reshape(image.shape))

plt.figure(figsize=(10,5))
plt.subplots_adjust(wspace=0.05, hspace=0.1)
plt.subplot(231)
plt.imshow(image)
plt.title("Original image")
plt.axis('off')

for idx, n_clusters in enumerate(n_colors):
    plt.subplot(232 + idx)
    plt.imshow(segmented_imgs[idx])
    plt.title("{} colors".format(n_clusters))
    plt.axis('off')
plt.show()
```
