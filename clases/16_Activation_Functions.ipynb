```python
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 14})
```

# Sigmoid function

```python
z = np.linspace(-10,10)
```

```python
f_z = 1/(1 + np.exp(-z))
```

```python
f_z
```

```python
def plot_act_func(label, title, z, f_z):
    plt.figure()
    plt.plot(z, f_z, label=label)
    plt.grid()
    plt.title(title)
    plt.xlabel("$z$")
    plt.ylabel("$f(z)$")
    plt.legend(loc="best")
    plt.show()
    
plot_act_func("$1/(1+e^{-z})$","Sigmoid activation function",z, f_z )
```

# Tanh

```python
f_z = np.tanh(z)
```

```python
plot_act_func("tanh", "Tanh", z, f_z)
```

```python
f_z =  (np.exp(z) - np.exp(-z))/(np.exp(z) + np.exp(-z))
```

```python
plot_act_func("tanh", "Tanh", z, f_z)
```

# ReLu

```python
f_z = np.maximum(0,z)
```

```python
plot_act_func("ReLu", "ReLu", z, f_z)
```
